```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nt = 30
r = 1/(2*np.pi)
ts = np.linspace(0, 2*np.pi, nt)
```

```python
def distance(ts):
    t1, t2 = np.expand_dims(ts, 0), np.expand_dims(ts, 1)
    diff = np.abs(t1 - t2) % (2 * np.pi)
    return np.where(diff > np.pi, 2 * np.pi - diff, diff) * r
```

```python
M = distance(ts)
print(np.max(M), " greater than ", 1/(2+1))
```

```python
plt.imshow(M)
plt.colorbar()
```

```python
A = np.expand_dims(M, 0) 
B = np.expand_dims(M, 1) 
C = np.expand_dims(M, 2) 

M = np.minimum(A, np.minimum(B, C)) 
print(np.max(M), " greater than ", 1/(3+1))
```
